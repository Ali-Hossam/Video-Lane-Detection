Print images

```python
def plot_images(img1, img2, label1="Origianl Image", label2 = "Cropped Image"):
    # Create a figure with two subplots
    fig, axes = plt.subplots(1, 2, figsize=(15, 5))

    # Plot the first image on the first subplot
    axes[0].imshow(img1, cmap='gray')
    axes[0].set_title(label1)

    # Plot the second image on the second subplot
    axes[1].imshow(img2, cmap='gray')
    axes[1].set_title(label2)

    # Show the plot
    plt.tight_layout()
    plt.show()
```

Crop (Region of interest) function

```python
def ROI(path):
    img = io.imread(path)
    d_row, d_col, _ = img.shape
    roi_maxr = d_row // 2
    # roi_maxc = d_col // 2
    print(d_row, roi_maxr)
    img_arr = np.array(img)
    if d_row >= 500:
        cropped_img = img_arr[(roi_maxr + 100):d_row,:]
    elif d_row > 350:
        cropped_img = img_arr[(roi_maxr + 40):d_row,:]
    else:
        cropped_img = img_arr[roi_maxr:d_row,:]
    return cropped_img
```

Smoothing using median filter

```python
def smooth(image):
    gray_image = rgb2gray(image)
    f_image = filters.median(gray_image)
    return f_image
```

Testing the function

```python
ROI1 = ROI('../images/fig12.gif')[:,:,:3]
ROI2 = ROI('../images/Horizontal-curves.jpg')[:,:,:3]
ROI3 = ROI('../images/istockphoto-954713540-612x612.jpg')[:,:,:3]
ROI4 = ROI('../images/photo-1539707313609-8f8bc5ec70e8.jpeg')[:,:,:3]
ROI5 = ROI('../images/road3.jpg')[:,:,:3]

show_images([io.imread('../images/fig12.gif') ,ROI1, smooth(ROI1)])
show_images([io.imread('../images/Horizontal-curves.jpg') ,ROI2, smooth(ROI2)])
show_images([io.imread('../images/istockphoto-954713540-612x612.jpg') ,ROI3, smooth(ROI3)])
show_images([io.imread('../images/photo-1539707313609-8f8bc5ec70e8.jpeg') ,ROI4, smooth(ROI4)])
show_images([io.imread('../images/road3.jpg') ,ROI5, smooth(ROI5)])
```
